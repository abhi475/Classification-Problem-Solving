THE MAIN IDEA OR OBJECTIVE OF THIS PROJECT IS: 
1. TO PREDICT THAT USER WILL PURCHASE THE PRODUCT OR NOT, BY TAKING INPUT(AGE AND ESTIMATED SALARY)
2. CHECKING THE ACCURACY OF THIS PROJECT BY USING DIFFERENT CLASSIFICATION ALGORITHM

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
data.head()
```

```python
data.iloc[:,2:4]
```

```python
data.iloc[:,2:4].values
```

```python
x=data.iloc[:,2:4].values
```

```python
x.shape
```

```python
data.iloc[:,-1]
```

```python
data.iloc[:,-1].values
```

```python
y=data.iloc[:,-1].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

1.BY USING KNN

```python
#CALCULATING VALUES OF K
np.sqrt(x_train.shape[0])
```

```python
k=16
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=k)
```

```python
#TRAIN OUR MODEL
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
def predict_output():
    age=int(input('ENTER YOUR AGE='))
    salary=int(input('ENTER YOUR SALARY='))
    
    x_new=np.array([[age],[salary]]).reshape(1,2)
    x_new=scaler.transform(x_new)
    
    if knn.predict(x_new)[0]==0:
        return "will not purchase"
    else:
        return "will purchase"
```

```python
predict_output()
```

2. BY USING DECISION TREE CLASSIFIER ALGORITHM

```python
from sklearn.tree import DecisionTreeClassifier
DTC= DecisionTreeClassifier(criterion='entropy', random_state=0)  
DTC.fit(x_train, y_train) 
```

```python
y_pred= DTC.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix  
confusion_matrix(y_test, y_pred) 
```

```python
from sklearn.metrics import precision_score
precision_score(y_test,y_pred)
```

```python
from sklearn.metrics import recall_score
recall_score(y_test,y_pred)
```

```python
from sklearn.metrics import f1_score
f1_score(y_test,y_pred)
```

```python
def predict_output():
    age=int(input('ENTER YOUR AGE='))
    salary=int(input('ENTER YOUR SALARY='))
    
    x_new=np.array([[age],[salary]]).reshape(1,2)
    x_new=scaler.transform(x_new)
    
    if DTC.predict(x_new)[0]==0:
        return "will not purchase"
    else:
        return "will purchase"
```
